```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=data.drop(["deck"],axis=1)
data.head(5)
```

```python
data.isnull().sum()
```

```python
age_mean=data.age.mean()
print(age_mean)

data.age=data.age.fillna(age_mean)
data.head(10)
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
plt.figure(figure=(4,2))

sns.histplot(data=data,x="age",hue="sex",kde=True)
plt.title("Age Distribution wrt Gender")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.barplot(data=data,x="class",y="survived",hue="who")
plt.title("Survival rate wrt passenger class and gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
sns.countplot(data=data,x="embarked",hue="survived")
plt.title("survival count based on embarkation port ")

data["family_size"]=data["sibsp"]+data["parch"]+1
plt.subplot(1, 2, 2)
sns.countplot(data=data,x="family_size",hue="survived")
plt.title("survival count based on family size")

plt.show()
```
